```python
# Downloading Data for January to December Using a For Loop

result_df = []


for i in range(1, 13):
    data = wget.download(f"https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2022-{i:02d}.parquet")
    df = pd.read_parquet(data)
    result_df.append(df)
final_df = pd.concat(result_df)
```

```python
# final_df is a concatenation of multiple files, I need to export it as a single file so it's in csv format
# This format would enable me send data in batches to postgresql 
# Writing to csv in order read it 

final_df.to_csv("green_taxi_dataset_jan_dec.csv", index=False)
```

```python
# Declaring a variable to read the above dataset into a pandas dataframe 

df_iter = pd.read_csv("green_taxi_dataset_jan_dec.csv", chunksize=100000)
```

```python
# Sending dataset in batches of 100000 using while loop

while True:

    time_start = time.time()
    
    df = next(df_iter)

    df.to_sql(name="test", con=engine, if_exists="append")

    time_end = time.time()

    print("inserted another chunk, this took %.3f seconds" % (time_end - time_start))
    
```
